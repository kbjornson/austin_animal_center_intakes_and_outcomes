## Data Cleaning and Preprocessing

### Transform 'DateTime' column to a datetime object

```python
# intakes
intakes['DateTime'] = pd.to_datetime(intakes['DateTime'], errors='coerce')
intakes.info()
```

```python
# outcomes
# pandas was having trouble converting to datetime, so remove the timezone offset
outcomes['DateTime'] = outcomes['DateTime'].str.replace(r'T', ' ', regex=True)
outcomes['DateTime'] = outcomes['DateTime'].str.replace(r'-\d{2}:\d{2}$', '', regex=True)
# convert to datetime object
outcomes['DateTime'] = pd.to_datetime(outcomes['DateTime'],errors='coerce')
outcomes.info()
```

### Check for Null Values
Although there are null values in both datasets, I am choosing to keep those records rather than remove them. For example, both datasets have nearly ~50k animals that do not have a name. This is often to be expected in a shelter situation, as shelters may intake animals without a prior known name. Excluding these animals could potentially skew the analysis, as these animals could disproportionally be strays, rather than intakes from a household. Additionally, I am choosing to keep records where other factors like "sex upon intake" and "age upon outcome" are null, because these are still valid records. I will treat these null values as a separate "unknown" group in Power BI visualizations.

### Check for Duplicates
Both datasets contain duplicates of Animal ID. For the intakes dataset, after looking at Animal IDs sorted by ID and date, it is evident that an animal can be intaken to the shelter multiple times throughout its life. Therefore these are valid records and not duplicates. Similarly, if an animal is taken in to the shelter multiple times, it also has multiple outcomes in the outcomes dataset. Therefore, these are valid records that can be kept in the dataset.

```python
# intakes
# count number of times each unique AnimalID appears in the dataset
animalid_counts = intakes['Animal ID'].value_counts()
# filter to AnimalIDs that appear more than once
duplicate_ids = animalid_counts[animalid_counts > 1]
print(duplicate_ids)
```

```python
# look at duplicated IDs sorted by ID and date
duplicate_rows = intakes[intakes['Animal ID'].isin(duplicate_ids.index)]
duplicates_sorted = duplicate_rows.sort_values(['Animal ID', 'DateTime'])
duplicates_sorted.head(10)
```

```python
# outcomes
# count number of times each unique AnimalID appears in the dataset
animalid_counts2 = outcomes['Animal ID'].value_counts()
# filter to AnimalIDs that appear more than once
duplicate_ids2 = animalid_counts2[animalid_counts2 > 1]
# look at duplicated IDs sorted by ID and date
duplicate_rows2 = outcomes[outcomes['Animal ID'].isin(duplicate_ids2.index)]
duplicates_sorted2 = duplicate_rows2.sort_values(['Animal ID', 'DateTime'])
duplicates_sorted2.head(10)
```

### Export to .csv for SQL
The next step is to export the cleaned datasets so they can be loaded into SQL. The plan is to use SQL to join the datasets to create a combined dataset of intakes and outcomes for each time an animal is taken in to the shelter.

```python
# export both dfs to .csv
intakes.to_csv('/Users/bjorn/Documents/Austin_Animal_Center_Project/Austin_Animal_Center_Intakes_Cleaned.csv', index=False, na_rep='NULL')
outcomes.to_csv('/Users/bjorn/Documents/Austin_Animal_Center_Project/Austin_Animal_Center_Outcomes_Cleaned.csv', index=False, na_rep='NULL')
```

```python
### SQL used to combine datasets

with intake_seq as (
	select *,
		row_number() over (partition by animal_id order by cast(datetime as date)) as seq
	from AnimalShelterDB.dbo.AnimalIntake
	),

outcome_seq as (
	select *,
		row_number() over (partition by animal_id order by date) as seq
	from AnimalShelterDB.dbo.AnimalOutcome
	)
		
select
	i.animal_id
	, i.name
	, i.animal_type
	, i.breed
	, i.color
	, cast(i.datetime as date) as intake_date
	, o.date as outcome_date
	, datediff(day, cast(i.datetime as date), o.date) as days_in_shelter
	, i.monthyear as intake_monthyear
	, o.monthyear as outcome_monthyear
	, i.intake_type
	, i.intake_condition
	, o.outcome_type
	, o.outcome_subtype
	, i.sex_upon_intake
	, o.sex_upon_outcome
	, case 
		when i.sex_upon_intake = 'Intact Male' and o.sex_upon_outcome = 'Neutered Male' then 'Y'
		when i.sex_upon_intake = 'Intact Female' and o.sex_upon_outcome = 'Spayed Female' then 'Y'
		else 'N'
	  end as neutered_at_shelter
	, i.age_upon_intake
	, o.age_upon_outcome
from intake_seq i
left join outcome_seq o
	on i.animal_id = o.animal_id
	and i.seq = o.seq
where o.date >= cast(i.datetime as date)
;
```
